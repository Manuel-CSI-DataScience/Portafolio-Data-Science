### Pregunta 5
Haz un histograma (`hist`) de la columna `ROM_GB` del **data frame** `df`.

El gráfico debe incluir un título y las columnas tener un color personalizado.

```python
import numpy as np
import matplotlib.pyplot as plt

# bins es rango de valores de los datos se divide en 65 intervalos para construir el histograma
# color ponemos el color que elegimos rojo
# Creamos el histograma
plt.hist(df["ROM_GB"], bins=65, label='GB Rom', color="red")

# Configurar el título y las etiquetas de los ejes
plt.title("Histograma de ROM en GB")
plt.xlabel("ROM en GB")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.legend()



# Mostrar el histograma
plt.show()
```

### Pregunta 6
Haz un diagrama de dispersión (`scatter`) con las siguientes características: el eje X deberá indicar el contenido de la columna `Precio de lista`, mientras que el eje Y expresará el contenido de la columna `Precio de venta`.

El diagrama debe incluir un título y los puntos tener un color personalizado.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Datos del diagrama de dispersión
x = df['Precio de lista']
y = df['Precio de venta']
color = 'green'  # Color personalizado para los puntos

# se crear el diagrama de dispersión scatter
plt.scatter(x, y, label='$Precio', color=color, marker='*')

# Configurar los ejes y el título
plt.xlabel('Precio de lista')
plt.ylabel('Precio de venta')
plt.title('Grafico de dispersión $Lista vs $Venta')
plt.grid(True)
plt.legend()
# el .show muestra el diagrama de dispersión
plt.show()
```

### Pregunta 10
¿Qué otro dato puedes obtener de `df` y cómo lo extraerías?

Podemos obtener los 10 celulares más calificados por los usuarios

```python
# Voy a obtener las lista de los 10 modelo de celulares mejor calificados
import numpy as np
import pandas as pd

# agrupamos las mejores calificaciones por modelo
mejores_calificaciones = df.groupby('Modelo')['Calificaciones'].max()

# ordeno las calificaciones
cali_orde = mejores_calificaciones.sort_values(ascending = False)
top_10 = cali_orde.head(10)

# Mostrar el resultado
print("nuestra lista de celulares más calificados es: \n")
top_10
```
